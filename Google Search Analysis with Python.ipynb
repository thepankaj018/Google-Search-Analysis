```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from pytrends.request import TrendReq
```

```python
trends=TrendReq()
```

```python
trends.build_payload(kw_list=['Machine Learning'])
```

```python
data=trends.interest_by_region()
```

```python
data=data.sort_values(by='Machine Learning',ascending=False).reset_index()
```

```python
data=data.head(10)
data
```

```python
plt.style.use('fivethirtyeight')
```

top 10 countries which search for “Machine Learning” on Google:

```python
plt.figure(figsize=(16,5))
sns.barplot(x='geoName',y='Machine Learning',data=data)
plt.show()
```

```python
data=TrendReq(hl='en-US', tz=360)
```

```python
data.build_payload(kw_list=['Machine Learning'])
```

```python
data = data.interest_over_time()
```

```python
data.head().reset_index()
```

```python
data.tail().reset_index()
```

```python
plt.figure(figsize=(20,15))
sns.lineplot(x='date',y='Machine Learning',data=data)
plt.title('Total Google Searches for Machine Learning', fontweight='bold')
plt.show()
```
